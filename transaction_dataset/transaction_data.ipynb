```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('transactions.csv')
df
```

```python
# Calculate total sales across all transactions
tot_sales = df['t_amt'].sum()
print(tot_sales)
```

```python
#which month had the highest total transaction amount?
df['t_date'] = pd.to_datetime(df['t_date'])
df['month'] = df['t_date'].dt.month
month_sales = df.groupby('month')['t_amt'].sum()
highest_month = month_sales.idxmax()
print(highest_month)

```

```python
#What is the average transaction amount per customer?
customer_avg = df.groupby('cust_id')['t_amt'].mean()
print(customer_avg)
```

```python
#What is the trend of total sales over months?
monthly_sales = df.groupby('month')['t_amt'].sum()
print(monthly_sales)
```

```python
#What is the highest single transaction amount recorded?
highest_amount = df['t_amt'].max()
print(highest_amount)
```

```python
#Which service category contributes the most to total revenue?
service_revenue = df.groupby('services')['t_amt'].sum()
top_service = service_revenue.idxmax()
print(top_service)
```

```python
#Which product generated the highest revenue?
product_revenue = df.groupby('products_used')['t_amt'].sum()
top_product = product_revenue.idxmax()
print(top_product)
```

```python
#How does the average transaction amount vary between services?
avg_transaction = df.groupby('services')['t_amt'].mean()
print(avg_transaction)
```

```python
#How many unique customers made purchases?
unique_customers = df['cust_id'].nunique()
print(unique_customers)
```

```python
#Which customers have spent the most overall?
cust_spending = df.groupby('cust_id')['t_amt'].sum()
top_customers = cust_spending.nlargest(3)
print(top_customers)
```

```python
#What is the average number of transactions per customer?
transactions_per_cust = df.groupby('cust_id').size()
avg_transactions = transactions_per_cust.mean()
print(avg_transactions)
```

```python
#Are there customers who made multiple purchases in different categories?
cust_categories = df.groupby(['cust_id', 'services']).size().reset_index(name='count')
multiple_category = cust_categories.groupby('cust_id').filter(lambda x: x['count'].sum() > 1)
print(multiple_category)
```

```python
#What percentage of customers are repeat buyers?
total_customers = df['cust_id'].nunique()
repeat_buyers = df.groupby('cust_id').size()
repeat_buyers_count = repeat_buyers[repeat_buyers > 1].count()
percentage_repeat_buyers = (repeat_buyers_count / total_customers * 100)
print(percentage_repeat_buyers)
```

```python
#Which product category has the highest total sales?
category_sales = df.groupby('services')['t_amt'].sum()
highest_sales_category = category_sales.idxmax()
print(highest_sales_category)
```

```python
#Which services are most popular by transaction count?
popular_services = df['services'].value_counts()
print(popular_services)
```

```python
#For each service, which product type is purchased most frequently?
service_product = df.groupby(['services', 'products_used']).size().reset_index(name='count')
most_frequent_product = service_product.loc[service_product.groupby('services')['count'].idxmax()]
print(most_frequent_product)
```

```python
#What is the average transaction amount per product type?
product_avg = df.groupby('products_used')['t_amt'].mean()
print(product_avg)
```

```python
#Are there specific services where customers spend significantly more?
service_avg = df.groupby('services')['t_amt'].mean()
overall_avg = df['t_amt'].mean()
significant_services = service_avg[service_avg > overall_avg]
print(significant_services)
```

```python
#Which state has the highest total sales?
state_sales = df.groupby('state')['t_amt'].sum()
top_state = state_sales.idxmax()
print(top_state)
```

```python
#Which city records the highest number of transactions?
city_transactions = df['city'].value_counts()
top_city = city_transactions.idxmax()
print(top_city)
```

```python
#What is the average spending per transaction in each state?
state_avg = df.groupby('state')['t_amt'].mean()
print(state_avg)
```

```python
#Are certain services more popular in specific states?
service_state = df.groupby(['services', 'state']).size().unstack(fill_value=0)
print(service_state)
```

```python
#Which states buy the most Outdoor Recreation products?
outdoor_sales = df[df['services'] == 'Outdoor Recreation'].groupby('state')['t_amt'].sum()
top_outdoor_state = outdoor_sales.idxmax()
print(top_outdoor_state)
```

```python
#Compare average spending between California and Texas customers.
ca_avg = df[df['state'] == 'California']['t_amt'].mean()
tx_avg = df[df['state'] == 'Texas']['t_amt'].mean()
print(f"Average spending in California: ${ca_avg:.2f}")
print(f"Average spending in Texas: ${tx_avg:.2f}")
```

```python
#Which quarter of the year saw the highest sales?
df['quarter'] = df['t_date'].dt.quarter
quarter_sales = df.groupby('quarter')['t_amt'].sum()
print(quarter_sales)
```

```python
#How does total sales vary month by month?
monthly_sales = df.groupby('month')['t_amt'].sum()
print(monthly_sales)
```

```python
#What is the total number of transactions per month?
transactions_per_month = df.groupby('month').size()
print(transactions_per_month)
```

```python
#How many transactions were done using credit?
credit_transactions = df[df['t_details'] == 'credit'].shape[0]
print(f"transactions done using credit: {credit_transactions}")
```

```python
#What is the total revenue generated from credit transactions?
credit_revenue = df[df['t_details'] == 'credit']['t_amt'].sum()
print(f"Total revenue from credit transactions: ${credit_revenue:.2f}")
```

```python
#Is there a difference in average spending between credit and debit customers?
avg_credit_spending = df[df['t_details'] == 'credit']['t_amt'].mean()
avg_debit_spending = df[df['t_details'] == 'debit']['t_amt'].mean()
difference = avg_credit_spending - avg_debit_spending
print(f"Difference: ${difference:.2f}")
```

```python
#Which states or cities should the company focus on for marketing high-value products?
state_revenue = df.groupby('state')['t_amt'].sum()
city_revenue = df.groupby('city')['t_amt'].sum()    
print(state_revenue.head())
print(city_revenue.head()) 
```

```python
#Should more inventory be stocked for Exercise & Fitness products given their popularity?
product_popularity = df.groupby('products_used')['t_amt'].sum()
print(product_popularity.sort_values(ascending=False))
```

```python
#Are there underperforming service categories that need promotional offers?
service_revenue = df.groupby('services')['t_amt'].sum()
print(service_revenue.sort_values())
```
